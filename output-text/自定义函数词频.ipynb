```python
import jieba
import jieba.posseg as pseg
```

```python
# 创建停用词list
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
```

```python
 # 对句子进行分词
def seg_sentence(sentence):
    sentence_seged = jieba.cut(sentence.strip())
    stopwords = stopwordslist('停用词.txt')
    seg_f = ''
    for word in sentence_seged:
        if word not in stopwords:
            seg_f += word
        else:
            seg_f += " "
    return seg_f
```

```python
#
def f_1(filepath):
    inputs = open(filepath, 'r', encoding='utf-8')
    outputs = ''
    for line in inputs:
        line_seg = seg_sentence(line)  # 这里的返回值是字符串
        outputs += (line_seg)
    fileee = open('file_f.txt', 'w', encoding='utf-8')
    fileee.write(outputs)
    fileee.close()
```

```python
def f_2():
    lines = []
    with open('file_f.txt','r',encoding='utf-8') as file:
        with open('file分词加词性.txt','w',encoding='utf-8') as segfile:
            for line in file:
                line = line.strip()
                if line != '':
                    lines.append(pseg.lcut(line))
                    for pair in pseg.lcut(line):
                        pair = list(pair)
                        segfile.write(pair[0] + '/' + pair[1] + ' ')
                    segfile.write('\n')
```

```python
def f_3():
    word_freq_2 = {}
    with open('file分词加词性.txt','r',encoding='utf-8') as segfile:
        for line in segfile:
            for pairs in line.split(' '):
                pair = pairs.split('/')
                if len(pair) > 1 and len(pair[0]) >=2 and pair[1] != 'v' :
                    if word_freq_2.get(pair[0]):
                        word_freq_2[pair[0]] += 1
                    else:
                        word_freq_2[pair[0]] = 1
    n = len(word_freq_2)
    word_freq_2_list = []
    for word in word_freq_2.keys():
        if word_freq_2[word] > 5:
            word_freq_2_list.append([word,word_freq_2[word]])
    sequence = sorted(word_freq_2_list, key=lambda item: item[1], reverse=True)
    print(n)
    return sequence
    
```

```python
def final(filepath):
    f_1(filepath)
    f_2()
    return f_3()
```

```python
#125规划词频
s_2 = final('125.txt')
print(s_2)
```

```python
#135规划词频
s_3 = final('135.txt')
print(s_3)
```

```python
only_125 = []
only_135 = []
only_145 = []
word_125 = [line.split(' ') for line in open('final_125词频.txt', 'r', encoding='utf-8').readlines()]
word_135 = [line.split(' ') for line in open('final_135词频.txt', 'r', encoding='utf-8').readlines()]
word_145 = [line.split(' ') for line in open('final_145词频.txt', 'r', encoding='utf-8').readlines()]

for w_2 in word_125[0]:
    if (w_2 not in word_135[0]) and (w_2 not in word_145[0]):
        only_125.append(w_2) 

for w_3 in word_135[0]:
    if (w_3 not in word_125[0]) and (w_3 not in word_145[0]):
        only_135.append(w_3) 


for w_4 in word_145[0]:
    if (w_4 not in word_125[0]) and (w_4 not in word_135[0]):
        only_145.append(w_4) 


share = []
for w_2 in word_125[0]:
    if (w_2 in word_135[0]) and (w_2 in word_145[0]):
        share.append(w_2) 
```

```python
#出现次数/总次数
def freq_num(sequence,sequence1,length):
    disc = {}
    for w in sequence:
        for word in sequence1:
            if w == word[0] and w not in disc:
                disc[w] = word[1]/length
            else:
                continue
    return disc
```

```python
#125规划分词出现次数/总次数
freq_num(share,s_2,len_2)
```
