```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
def getHTML(u):
    try:
        r = requests.get(u,timeout=30)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return 

def getContent(u):
    html = getHTML(u)
    soup = BeautifulSoup(html,'html.parser')
    title = soup.select('div.mbtitle')
    paras_tmp = soup.select('p')
    paras = paras_tmp[3:]
    return paras
 
def saveFile(text):
    f=open('try.txt','w',encoding='utf-8')
    for t in text:
        if len(t) > 0:
            f.writelines(t.get_text() + "\n")
    f.close()
    
def finish(u):
    text = getContent(u)
    saveFile(text)
    
def web(u):
    headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.146 Safari/537.36'
    }
    url = u
    res = requests.get(url, headers=headers)
    res.encoding='gbk'
    soup = BeautifulSoup(res.text, 'lxml')

    news_list = soup.find('div', {'class': 'pages_content'})
    if news_list==None:
        for x in ['wrap','zw-con']:
            news_list = soup.find('div', {'class': x})
            if news_list!=None:
                break
    news = news_list.find_all('p')

    txt=open('try.txt','w')
    for x in news:
        a=x.text
        txt.write(a)
```
