```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#读取csv
df = pd.read_csv('全国结婚离婚数据2.csv',encoding='gbk')
df = df.drop([0,1])
print(df)
```

```python
print(df.iloc[1])
```

```python
print(df.iloc[1][1])
```

```python
marriage = 0
divorce = 0
m_vertical = []
d_vertical = []
abscissa = []
for i in range(1,111,2):
    marriage += int(df.iloc[1][i])
    if (i+1)%8 == 0:
        m_vertical.append(marriage)
        continue
        
for i in range(2,111,2):
    divorce += int(df.iloc[1][i])
    if i%8 == 0:
        d_vertical.append(divorce)
        continue
        
        
for j in range(2007,2020):
    abscissa.append(j)
    
print(d_vertical)
print(m_vertical)
print(abscissa)
```

```python
fig,ax = plt.subplots()
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.xlabel('年/(年)')
plt.ylabel('数量/(对)')
 

yticks = range(50000,5000000,550000)
ax.set_yticks(yticks)

xticks = range(2007,2020,1)
ax.set_xticks(xticks)
 
ax.set_ylim([50000,5000000])
ax.set_xlim([2007,2019])

x = abscissa
plt.plot(x,m_vertical,"x-",label="结婚数")
plt.plot(x,d_vertical,"+-",label="离婚数")
plt.title('全国结婚和离婚数量统计')

plt.grid(True)


for a,b in zip(x,m_vertical):
    plt.text(a, b+1, '%.0f' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,d_vertical):
    plt.text(a, b+1, '%.0f' % b, ha='center', va= 'bottom',fontsize=9)


    
plt.legend(bbox_to_anchor=(1.0, 1), loc=2, borderaxespad=-5.)
 
plt.show()
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

x = np.arange(13)
y = m_vertical
y1 = d_vertical

bar_width = 0.35
tick_label = abscissa

plt.bar(x, y, bar_width, align="center", label="结婚数", alpha=0.5)
plt.bar(x+bar_width, y1, bar_width, align="center", label="离婚数", alpha=0.5)
for a,b in zip(x,y):
    plt.text(a, b+1, '%.0f' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,y1):
    plt.text(a, b+1, '%.0f' % b, ha='center', va= 'bottom',fontsize=7)

plt.xlabel("年份")
plt.ylabel("数量")

plt.xticks(x+bar_width/2, tick_label)

plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def policy_num():
    df = pd.read_excel(r'C:\Users\HONEYMOON\Desktop\规划\中央创业政策2.xlsx')
    s = df['发布时间'].dt.date.values
    year = {}
    for i in s:
        if str(i)[:4] not in year:
            year[str(i)[:4]] = 1
        else:
            year[str(i)[:4]] += 1


    num_lst = []
    year_lst = []
    for i in year.values():
        num_lst.append(i)
    for i in year.keys():
        year_lst.append(i)
    num_lst = num_lst[::-1]
    year_lst = year_lst[::-1]


    x = year_lst
    y = num_lst
    plt.plot(x, y,marker='o',linestyle='dashed')
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.xlabel('年/(年)')
    plt.ylabel('数量/(份)')
    plt.title('各年份创业政策数量统计图')
    for a,b in zip(x,y):
        plt.text(a, b+1, '%.0f' % b, ha='center', va= 'bottom',fontsize=12)
    plt.show()
```
