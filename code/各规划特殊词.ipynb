```python
only_125 = []
only_135 = []
only_145 = []
word_125 = [line.split(' ') for line in open('final_125词频.txt', 'r', encoding='utf-8').readlines()]
word_135 = [line.split(' ') for line in open('final_135词频.txt', 'r', encoding='utf-8').readlines()]
word_145 = [line.split(' ') for line in open('final_145词频.txt', 'r', encoding='utf-8').readlines()]

for w_2 in word_125[0]:
    if (w_2 not in word_135[0]) and (w_2 not in word_145[0]):
        only_125.append(w_2) 
print("125规划中特有的词汇：",only_125)
print()

for w_3 in word_135[0]:
    if (w_3 not in word_125[0]) and (w_3 not in word_145[0]):
        only_135.append(w_3) 
print("135规划中特有的词汇：",only_135)
print()

for w_4 in word_145[0]:
    if (w_4 not in word_125[0]) and (w_4 not in word_135[0]):
        only_145.append(w_4) 
print("145规划中特有的词汇：",only_145)
print()

share = []
for w_2 in word_125[0]:
    if (w_2 in word_135[0]) and (w_2 in word_145[0]):
        share.append(w_2)
print("三个规划中共有的词汇：",share)
print()
```
