```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
frequencies = {}
for line in open("final_125词频2.txt",encoding='utf-8'):
    arr = line.split(" ")
    arr[1] = arr[1].replace("\n", "") 
    print(arr)
    frequencies[arr[0]] = float(arr[1])
wc = WordCloud(
    max_font_size = 300,
    colormap = "Reds",
    font_path="fonts/msyh.ttc",
    max_words=300,
    width=2000,
    height=1200,
)
word_cloud = wc.generate_from_frequencies(frequencies)
# 写词云图片
word_cloud.to_file("wc_125.jpg")
# 显示词云文件
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def wordcloud(filepath):
    frequencies = {}
    for line in open(filepath,encoding='utf-8'):
        arr = line.split(" ")
        arr[1] = arr[1].replace("\n", "") 
        print(arr)
        frequencies[arr[0]] = float(arr[1])
    wc = WordCloud(
        max_font_size = 300,
        colormap = "Reds",
        font_path="fonts/msyh.ttc",
        max_words=300,
        width=2000,
        height=1200,
    )
    word_cloud = wc.generate_from_frequencies(frequencies)
    # 写词云图片
    word_cloud.to_file("wc.jpg")
    # 显示词云文件
    plt.imshow(word_cloud)
    plt.axis("off")
    plt.show()
```

```python
wordcloud("final_125词频2.txt")
```
